## üí° 1Ô∏è What is TensorFlow?

TensorFlow is an open-source framework developed by Google Brain for building and training machine learning (ML) and deep learning (DL) models.

The name breaks down like this:

- Tensor ‚Üí a mathematical object (a multi-dimensional array) used to represent data.

- Flow ‚Üí the way these tensors move through a computational graph.

So TensorFlow means **flow of tensors through a graph of operations.**

It can run on:

- CPUs and GPUs (for fast training)

- TPUs (Tensor Processing Units)

- desktops, servers, mobile devices, and the cloud.

---

## üîÆ 2Ô∏è How TensorFlow Works (The Core Idea)

Data enters the model as tensors (e.g., images, text, numbers).

TensorFlow creates a computational graph describing the flow of data and operations.

Each operation (e.g., addition, matrix multiplication, activation) is a node in the graph.

During training, TensorFlow uses automatic differentiation to compute gradients (derivatives).

These gradients update the model‚Äôs weights through backpropagation to minimize the loss function.

üí¨ Think of it like a smart calculator that builds a map of how every operation connects, so it can automatically figure out how to adjust everything to learn patterns.

## üß† 3Ô∏è TensorFlow Architecture at a Glance

TensorFlow‚Äôs core design is built around the idea of **tensors** flowing through a **graph of operations**.  
Each component plays a specific role in how data is processed and learned by the model.

| üß© **Component** | üß† **Role** |
|------------------|-------------|
| **Tensors** | The basic data units ‚Äî multi-dimensional arrays (like NumPy ndarrays) that hold numerical data. |
| **Operations (Ops)** | The functions that transform tensors (e.g., addition, multiplication, activation functions). |
| **Graph** | A computational map that connects operations and tensors ‚Äî defines how data moves through the network. |
| **Session (TF 1.x)** | Used to execute the graph in older versions of TensorFlow (replaced by eager execution in TF 2.x). |
| **Eager Execution (TF 2.x)** | Runs operations immediately without building a graph first ‚Äî makes TensorFlow feel like regular Python. |
| **Keras API** | A user-friendly, high-level interface for building, training, and evaluating neural networks. |

## üîÅ 6Ô∏è Automatic Differentiation (GradientTape)

Training neural networks means finding how changing weights affects the loss.  
TensorFlow‚Äôs **`tf.GradientTape`** automatically records operations and computes derivatives for you.

**Concept example**

\[
y = x^2 + 3x + 5 \quad \Rightarrow \quad \frac{dy}{dx} = 2x + 3
\]

TensorFlow performs this process on thousands of parameters using **backpropagation**.

## üìâ 8Ô∏è The Training Loop (Under the Hood)

Each training epoch performs:

1. **Forward pass** ‚Äî generate predictions  
2. **Compute loss** ‚Äî measure prediction error  
3. **Backward pass** ‚Äî calculate gradients with `GradientTape`  
4. **Optimizer step** ‚Äî update weights  
5. Repeat until convergence üåÄ  

This process allows the model to continuously improve.

#### 2Ô∏è Creating and Exploring Tensors
Tensors are multi-dimensional arrays ‚Äî the building blocks of TensorFlow

````python
import tensorflow as tf
import numpy as np

# Scalars, vectors, and matrices
scalar = tf.constant(5)
vector = tf.constant([1, 2, 3])
matrix = tf.constant([[1, 2], [3, 4]])
tensor3 = tf.constant([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])

print("Scalar:", scalar.numpy())
print("Vector:", vector.numpy())
print("Matrix:\n", matrix.numpy())
print("3-D Tensor shape:", tensor3.shape)
````

````python
# Tensor properties
print("Datatype:", tensor3.dtype)
print("Shape:", tensor3.shape)
print("Rank:", tf.rank(tensor3).numpy())
print("Number of dimensions:", tensor3.ndim)
````

#### 3Ô∏è Basic Tensor Operations
TensorFlow provides vectorized math operations just like NumPy ‚Äî but optimized for GPU.


````python
a = tf.constant([[1, 2], [3, 4]])
b = tf.constant([[5, 6], [7, 8]])

print("Add:\n", tf.add(a, b).numpy())
print("Multiply:\n", tf.multiply(a, b).numpy())
print("Matrix Multiplication:\n", tf.matmul(a, b).numpy())

# Broadcasting
x = tf.constant([1, 2, 3])
y = tf.constant(2)
print("Broadcast multiply:", (x * y).numpy())
````

#### 4Ô∏è Using Variables
`tf.Variable` lets you update values (useful for model weights).


````python
w = tf.Variable(5.0)
b = tf.Variable(1.0)

with tf.GradientTape() as tape:
    y = w**2 + 3*w + b
grad = tape.gradient(y, [w, b])

print("Value of y:", y.numpy())
print("Gradients:", [g.numpy() for g in grad])
````

####  5Ô∏è Simple Linear Regression with TensorFlow
We‚Äôll train a model to learn the relation **y = 3x + 2**

````python
# Generate data
X = tf.constant(np.linspace(-10, 10, 100), dtype=tf.float32)
Y = 3*X + 2 + tf.random.normal(shape=[100], mean=0, stddev=1)

# Initialize parameters
W = tf.Variable(tf.random.normal([1]))
B = tf.Variable(tf.random.normal([1]))
lr = 0.01

# Training loop
for epoch in range(300):
    with tf.GradientTape() as tape:
        Y_pred = W*X + B
        loss = tf.reduce_mean(tf.square(Y - Y_pred))
    dW, dB = tape.gradient(loss, [W, B])
    W.assign_sub(lr*dW)
    B.assign_sub(lr*dB)

print("Learned parameters:", W.numpy(), B.numpy())
````

````python
import matplotlib.pyplot as plt

plt.scatter(X, Y, label="True Data", color='skyblue')
plt.plot(X, W*X + B, color='red', label='Model Prediction')
plt.legend()
plt.title("Simple Linear Regression ‚Äî TensorFlow")
plt.show()
````

#### 6Ô∏è Building Your First Neural Network (Keras API)
We‚Äôll use MNIST digits to train a simple classifier.


````python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

# Load data
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train/255.0, x_test/255.0

# Build model
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train
history = model.fit(x_train, y_train, epochs=5,
                    validation_split=0.1, verbose=1)
````

````python
# Evaluate
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f" Test Accuracy: {test_acc:.4f}")
````

````python
# Plot training curves
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Val')
plt.title('Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
````

````python
#7Ô∏è Saving and Loading Models

model.save("basic_mnist_model.h5")
loaded_model = tf.keras.models.load_model("basic_mnist_model.h5")
print("Model reloaded and ready for inference!")
````
